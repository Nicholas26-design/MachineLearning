## How to make random numbers

```python
one_d = np.random.rand(5)
one_d
two_d = np.random.rand(5, 5)
two_d
```

```python
# Random numbers from a standard dist

one_d = np.random.randn(4)
one_d
two_d = np.random.randn(4, 4)
two_d
```

```python
# Random integers from a low to a high number

rand = np.random.randint(4, 51, 10)
rand
```
