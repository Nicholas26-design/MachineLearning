```python
import numpy as np
```

```python
arr = np.arange(0, 13)
```

```python
arr
```

## How to pick something out of array

```python
arr[9]
```

```python
arr[0: 7] # everything between two points
```

```python
arr[:7] # everything to the defined point
```

```python
arr[5:] # everything from and after defined point
```

```python
# You can set values in an array
arr[0:6] = 100
```

```python
arr
```

```python
# Reset
arr = np.arange(0, 13)
```

```python
# Slice/subset elements of the array into its own array
slice_of_arr = arr[0:6]
slice_of_arr[:] #pulls all of the array
```

```python
# If you change values of the slice, what happens to original?
slice_of_arr[:] = 50
slice_of_arr
arr

# slice_of_arr is not a copy of the original array. Therefore changes you make, change the original.
```

```python
# Reset
arr = np.arange(0, 13)
```

```python
## If you want to copy original and change in isolation, do the following:

arr_copy = arr.copy()
slice2 = arr_copy[0:5]
slice2[:] = 50
slice2
arr_copy
arr
```
