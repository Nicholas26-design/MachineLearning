```python
# How many people have the word Chief in their job title? I didn't know this.

def chief_string(title):
    if 'chief' in title.lower():
        return True
    else:
        return False
```

```python
sum(sal['JobTitle'].apply(lambda x: chief_string(x)))
```

Another struggle point:
