## Multi-Index and Index Hierarchy

```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside)) # zipped it together and made it a list type. It's like dictionary format.
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df = pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
df
# 
```

### Call data from this multi index. 

```python
# If we want to call everything from the first index
df.loc['G1']
```

```python
# The method is outside, then in. So information at G1, position 2 is:

df.loc['G1'].loc[2]
```

```python
## Giving names

df.index.names = ['Group','Num']
```

```python
# Returning a certain value
df.loc['G2'].loc[2]['B']
```
