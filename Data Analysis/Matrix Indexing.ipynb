## Indexing a Matrix

```python
matrix = np.array([[5,10,15], [20,25,30], [35,40,45]])
```

```python
matrix
```

```python
# If I wanted to grab 5 from matrix
matrix[0] [0] # row, column
# If I wanted the first row of matrix
matrix[1]
# If I wanted the first column of matrix?
```

```python
# comma notation
matrix[0,1]
matrix[1]
```

```python
# Navigating how much of the array you want requires slice notation.
# If I want the top 4 elements from the right:
matrix[:2, 1:]
```

```python
# I want just middle column
matrix[:3]
```

## The above is pretty important just in terms of slice notation and using brackets to navigate. 

## Important: Conditional Selection

```python
arr = np.arange(1, 11)
```

```python
arr
```

```python
# applying a conditional to your array.
bool_arr = arr > 5
bool_arr
```

```python
# Conditonal selection
arr[arr < 4]
```

```python
arr_2d = np.arange(50).reshape(5, 10)
arr_2d
```

```python
# Indexing Practice
arr_2d[1:, 5:]
```
