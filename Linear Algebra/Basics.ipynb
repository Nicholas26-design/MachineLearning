Information on vectors, arrays, and matricies

```python
import numpy as np
import pandas as pd
```

A vector is an ordered sequence of numbers and the order matters. Vectors can be any size. No constraint. Up or down.
A matrix is a collection of vectors with the same number of elements. 
Really you can classify a matrix of m rows and m columns as a collection of vectors, but the distinction is that a matrix has the 2D structure and conceptually, a collection of vectors don't posses a fixed structure. 

a matrix with n columns can be interpreted as a collection of n column vectors, each with the same number of elements. Similarly, a matrix with m rows can be interpreted as a collection of m row vectors.

In relation to machine learning, a vector is a row in a dataframe.
In relation to machine learning, a matrix is the dataframe.

On the other hand, a vector, in the context of data processing, can be thought of as a specific type of array that represents a sequence of values. Typically, a vector in Python is represented using a one-dimensional array or a specialized data structure like the NumPy array.

Information and examples of Arrays:

Arrays are similar to matrices in that they are also rectangular collections of elements. However, arrays are more general and can have any number of dimensions. 
They can be one-dimensional (like a list), two-dimensional (like a matrix), three-dimensional, and so on. 

An array is a data structure that can store a general collection of elements. Element type can vary. 

On the other hand, a vector, in the context of data processing, can be thought of as a specific type of array that represents a sequence of values. Typically, a vector in Python is represented using a one-dimensional array or a specialized data structure like the NumPy array.
